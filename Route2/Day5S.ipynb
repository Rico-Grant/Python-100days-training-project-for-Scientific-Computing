题目描述

请编写一个函数 reverse_string_with_stack(s)。
你要利用 栈 (Stack) 的思想，把一个字符串进行反转。

逻辑提示：

初始化一个空列表作为“栈”。

入栈阶段：遍历字符串，把每个字符逐一 .append() 进栈里。

出栈阶段：只要栈不为空，就不断地调用 .pop() 弹出字符，并把它们拼接在一起。

返回拼接后的新字符串。

```python
def reverse_string_with_stack(s):
    stack = []
    for char in s:
        stack.append(char)
    reversed_string = ''

    while stack:
        reversed_string += stack.pop()
    return reversed_string
# Example usage
input_string = "Hello, World!"
reversed_string = reverse_string_with_stack(input_string)
print(reversed_string)  # Output: !dlroW ,olleH
```

```python
from collections import deque

def simulate_checkout(customers):
    queue = deque(customers)
    while queue:
        current_customer = queue.popleft()
        print(f"正在为{current_customer}结账...")


n = 5    
customers = []
while len(customers) < n:
    customer_name = input("Enter customer name: ")
    customers.append(customer_name)

simulate_checkout(customers)



```
