题目描述

请你编写一个函数 count_greater_than(nums, threshold)。
该函数接收两个参数：

    nums：一个包含整数的列表。

    threshold：一个整数阈值。

你的任务是统计并返回列表中所有大于该阈值的数字个数。

```python
def count_greater_than(nums , threshold):
    count = 0
    for num in nums:
        if num > threshold:
            count += 1
    return count

list1 = [1, 5, 3, 8, 2]
threshold = 4
result = count_greater_than(list1, threshold)
print(f"Number of elements greater than {threshold}: {result}")
```

题目描述

请编写一个函数 find_first_index(nums, target)。
该函数接收两个参数：

    nums：一个整数列表。

    target：我们要寻找的目标整数。

你的任务是：找到目标数字 target 在列表中第一次出现的下标（索引），并将其返回。

    如果找到了，直接返回下标。

    如果遍历完整个列表都没找到，请返回 -1。

输入输出要求

    输入：一个整数列表和一个整数。

    输出：对应的下标（整数）或 -1。

```python
def find_first_index(nums, target):
    for index in range(len(nums)):
        if nums[index] == target:
            return index
    return -1

list2 = [10, 20, 30, 40, 50]
target = 30
index_result = find_first_index(list2, target)
print(f"First index of {target}: {index_result}")
```

```python
def find_max_and_index(nums):
    i = 0
    max_value = int(nums[0])
    max_index = 0
    while i < len(nums):
        if nums[i] > max_value:
            max_value = nums[i]
            max_index = i
        i += 1
    return [max_value, max_index]

list3 = [3, 7, 2, 9, 5]
max_value, max_index = find_max_and_index(list3)
print(f"Maximum value: {max_value}, Index of maximum value: {max_index}")

```

```python
def is_palindromes(s):
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

s = 'racecar'
print(is_palindromes(s))
```

你的任务是：找出同时出现在这两个列表中的数字，并以 列表 的形式返回。要求结果中不含重复数字。

```python
import re


def find_common_elemenets(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    res = []
    for i in s1:
        if i in s2:
            res.append(i)
    return res


list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8] 
common_elements_result = find_common_elemenets(list1, list2)
print(f"Common elements: {common_elements_result}")
```

```python
def bubble_sort(nums):
    for i in range(len(nums)):
        # 冒泡排序的内层循环，每次比较相邻的元素，如果顺序错误就交换它们
        # 最大的元素会在每次外层循环结束时被放到正确的位置
        # 内层循环的范围逐渐减少，因为每次外层循环结束后，最后的元素已经是正确的了
        for j in range(0, len(nums) - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
            

nums = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(nums)
print(f"Sorted array: {nums}")

```

题目描述

请编写一个函数 binary_search(nums, target)。
该函数接收一个 已经从小到大排好序 的整数列表 nums 和一个目标值 target。

你的任务是：使用 二分查找 算法找到 target 的下标。
输入输出要求

输入：有序列表 [1, 3, 5, 7, 9]，目标值 7。

输出：下标 3。

未找到：返回 -1。

```python
def bubble_sort(nums):
    for i in range(len(nums)):
        # 冒泡排序的内层循环，每次比较相邻的元素，如果顺序错误就交换它们
        # 最大的元素会在每次外层循环结束时被放到正确的位置
        # 内层循环的范围逐渐减少，因为每次外层循环结束后，最后的元素已经是正确的了
        for j in range(0, len(nums) - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
            
def Binary_search(nums, target):
    low = 0
    high = len(nums)-1
    while low <= high:
        mid = (low + high) // 2
        if nums[mid] == target:
            return mid
        else:
            if nums[mid] < target:
                low = mid + 1
            else:
                high = mid - 1
    return -1


nums = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(nums)
target = 25
index_result = Binary_search(nums, target)
print(f"Index of {target}: {index_result}")



        
```

题号：013

题型：递归基础练习

考察点：递归基准情况、递归公式实现

难度：入门级 🌟🌟🌟

题目描述

请编写一个递归函数 factorial(n)。
该函数接收一个正整数 n，计算并返回它的 阶乘（即 n!=n×(n−1)×⋯×1）。

递归公式提示：

基准情况：如果 n=1，返回 1。

递归步骤：n 的阶乘 = n×(n−1) 的阶乘。
即：return n * factorial(n - 1)

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
number = 5
factorial_result = factorial(number)
print(f"Factorial of {number}: {factorial_result}")
```

题号：014

题型：递归算法进阶

考察点：多重递归、多基准情况处理

难度：入门级 🌟🌟🌟+

题目描述

请编写一个递归函数 fibonacci(n)。
该函数接收一个非负整数 n，返回斐波那契数列中第 n 个位置的数字（假设下标从 0 开始）。

逻辑提示：

基准情况 1：如果 n=0，返回 0。

基准情况 2：如果 n=1，返回 1。

递归步骤：返回 fibonacci(n - 1) + fibonacci(n - 2)。

示例
Python

print(fibonacci(6)) 
0, 1, 1, 2, 3, 5, 8... 
第 6 个数是 8

作答要求

必须使用递归。

注意处理两个基准情况。

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
print(fibonacci(40)) 
```
