# 第35题 简谐运动

$$x(t)=A\cos(\omega t + \phi)$$

````python
import math
class SHM:
    def __init__(self, A, T, phi):
        self.A = A
        self.T = T
        self.phi = phi
    def frequency(self):
        return self.T ** -1
    def omega(self):
        return 2 * math.pi * self.frequency()
    def displacement(self, t):
        return self.A * math.sin(self.omega() * t + self.phi)
    def show_info(self,t):
        print(f"f = {self.frequency()}")
        print(f"omega = {self.omega()}")
        print(f"x = {self.displacement(t)}")
shm = SHM(float(input()), float(input()), math.radians(float(input())))
shm.show_info(float(input()))
````

## 题目：行星引力计算器

对于质量为 $M$，半径为 $R$ 的星球，设万有引力常量：$
G = 6.67 \times 10^{-11}
$

有以下公式：

### 1. 表面重力加速度
$$
g = \frac{GM}{R^2}
$$

### 2. 第一宇宙速度（环绕表面附近圆轨道速度）
$$
v_1 = \sqrt{\frac{GM}{R}}
$$

### 3. 第二宇宙速度（逃逸速度）
$$
v_2 = \sqrt{\frac{2GM}{R}}
$$

---



````python
import math

class Planet:
    def __init__(self,M,R):
        self.M = M
        self.R = R
    def surface_gravity(self):
        G = 6.67430e-11
        return G * self.M / self.R ** 2
    def first_cosmic_velocity(self):
        G = 6.67430e-11
        return math.sqrt(G * self.M / self.R)
    def second_cosmic_velocity(self):
        G = 6.67430e-11
        return math.sqrt(2 * G * self.M / self.R)
    def show_info(self):
        print(f"g = {self.surface_gravity()}")
        print(f"v1 = {self.first_cosmic_velocity()}")
        print(f"v2 = {self.second_cosmic_velocity()}")

planet = Planet(float(input()), float(input()))
planet.show_info()

    
````

## 题目：二维力分解模型

一个力F与x轴正方向夹角为θ(单位：度)，其分量为：

$$ F_{x}=F\cos (\theta) $$

$$ F_{y}=F\sin (\theta) $$

合力大小验证公式：

$$ |\vec{F}|=\sqrt{F_{x}^{2}+F_{y}^{2}} $$

````python
import math

class Force2D:
    def __init__(self,F,theta):
        self.F = F
        self.theta = math.radians(theta)
    def F_x(self):
        return self.F * math.cos(self.theta)
    def F_y(self):
        return self.F * math.sin(self.theta)
    def magnitude_check(self):
        return math.sqrt(self.F_x() ** 2 + self.F_y() ** 2)
    def show_info(self):
        print(f"Fx = {self.F_x()}")
        print(f"Fy = {self.F_y()}")
        print(f"F_check = {self.magnitude_check()}")

force = Force2D(float(input()), float(input()))
force.show_info()
````

# 第38题：电容充电模型

一个RC电路中，电容充电过程满足：

$$ Q(t)=CV\left(1-e^{-t /(RC)}\right) $$

$$ I(t)=\frac{V}{R} e^{-t /(R C)} $$

其中：
- $ Q(t) $：时刻 $ t $ 的电荷量
- $ I(t) $：时刻 $ t $ 的电流
- $ R $：电阻
- $ C $：电容
- $ V $：电源电压

````python
import math

class RCCharge:
    def __init__(self,R,C,V):
        self.R=R
        self.C=C
        self.V=V
    def tau(self):
        return self.R * self.C
    def charge(self,t):
        return self.C * self.V * (1-math.exp(-t/(self.R * self.C)))
    def current(self,t):
        return (self.V/self.R)*(math.exp(-t/(self.R*self.C)))
    def show_info(self,t):
        print(f"tau = {self.tau()}")
        print(f"Q = {self.charge(t)}")
        print(f"I = {self.current(t)}")
    
rc = RCCharge(float(input()),float(input()),float(input()))
rc.show_info(float(input()))
````

题目：二维速度合成模型

一个质点同时具有两个互相垂直的速度分量$ v_{x} $和$ v_{y} $。

总速度大小为：

$$ v = \sqrt{v_{x}^{2} + v_{y}^{2}} $$

速度方向角（相对x轴正方向）为：

$$ \theta = \arctan\left(\frac{v_{y}}{v_{x}}\right) $$

注意：

- 方向角最终输出为度
- 应使用math atan2(vy,vx)，这样更安全

````python
import math

class Velocity2D:
    def __init__(self,vx,vy):
        self.vx = vx
        self.vy = vy
    def speed(self):
        return math.sqrt(self.vx ** 2 + self.vy ** 2)
    def angle(self):
        return math.degrees(math.atan2(self.vy, self.vx))
    def show_info(self):
        print(f"speed = {self.speed()}")
        print(f"angle = {self.angle()}")
velocity = Velocity2D(float(input()), float(input()))
velocity.show_info()
````

题目：弹簧振子能量模型

一个质量为$m$的物体连接在劲度系数为$k$的弹簧上，振幅为$A$。在简谐振动中：

1.  总机械能
    $$E = \frac{1}{2}kA^2$$

2.  某位移$x$处的弹性势能
    $$U = \frac{1}{2}kx^2$$

3.  某位移$x$处的动能
    $$K = E - U$$

4.  某位移$x$处的速度大小
    $$v = \sqrt{\frac{2K}{m}}$$

````python
import math
class SpringSHM:
    def __init__(self,m,K,A):
        self.m = m
        self.k = K
        self.A = A
    def total_energy(self):
        return 0.5 * self.k * self.A ** 2
    
    def poential_energy(self,x):
        return 0.5 * self.k * x ** 2
    
    def kinetic_energy(self,x):
        return self.total_energy() - self.poential_energy(x)
    
    def speed(self,x):
        return math.sqrt(2 * self.kinetic_energy(x) / self.m)
    
    def show_info(self,x):
        print(f"E = {self.total_energy()}")
        print(f"U = {self.poential_energy(x)}")
        print(f"K = {self.kinetic_energy(x)}")
        print(f"v = {self.speed(x)}")
        
spring = SpringSHM(float(input()), float(input()), float(input()))
spring.show_info(float(input()))
````

题目：自由落体数据表生成器

一个物体从高度$ h_{0} $以初速度$ v_{0} $竖直向上运动，忽略空气阻力，加速度为：

$$ a = -g $$

其中$ g = 9.8 $。

位移公式：

$$ y(t) = h_0 + v_0 t - \frac{1}{2}gt^2 $$

速度公式：

$$ v(t) = v_0 - gt $$

任务：定义一个类 VerticalMotion

````python
class VerticalMotion:
    def __init__(self,h0,v0):
        self.h0 = h0
        self.v0 = v0
        self.g = 9.8
    def height(self,t):
        return self.h0 + self.v0 * t - 0.5 * self.g * t ** 2
    def velocity(self,t):
        return self.v0 - self.g * t
    def generate_table(self,t_end,dt):
        t = 0
        table = []
        while t <= t_end:
            table.append((t,self.height(t),self.velocity(t)))
            t += dt
        return table
    def show_table(self,t_end,dt):
        table = self.generate_table(t_end,dt)
        for t,h,v in table:
            print(f"t = {t}, h = {h}, v = {v}")

vm = VerticalMotion(float(input("h_0")),float(input("v_0")))
vm.generate_table(float(input("t_end")),float(input("dt")))
````

**题目：自由落体的数值迭代模拟**

一个物体从高度 $h_0$ 以初速度 $v_0$ 竖直运动，重力加速度取：

$$ g = 9.8 $$

现在不要使用解析公式，而要用数值迭代方法模拟。

**数值更新规则**

每经过一个时间步长 $dt$，按以下规则更新：

$$ v_{\text{new}} = v - g \cdot dt $$
$$ h_{\text{new}} = h + v \cdot dt $$

**注意：**

- 先更新速度还是先更新位置，要严格按题目给的规则理解。
- 这里的位置更新使用的是旧速度 $v$。

````python
class FallSimulation:
    def __init__(self, h0, v0):
        self.h0 = h0
        self.v0 = v0
        self.g = 9.8

    def simulate(self, dt):
        t = 0.0
        h = self.h0
        v = self.v0
        data = [(t, h, v)]

        while h >= 0:
            v_new = v - self.g * dt
            h_new = h + v * dt
            t += dt
            data.append((t, h_new, v_new))
            v = v_new
            h = h_new

            if h < 0:
                break

        return data

    def show_data(self, dt):
        data = self.simulate(dt)
        for t, h, v in data:
            print(f"t = {t}, h = {h}, v = {v}")

fs = FallSimulation(float(input()), float(input()))
fs.show_data(float(input()))
````

第43题

题目：抛体运动数据表+最高点查找

继续做一个更像建模的小题。

一个物体以初速度$ v_{0} $、发射角$ \theta $从地面发射，忽略空气阻力，重力加速度$ g=9.8 $。

公式

把初速度分解为：

$$ v_{x 0}=v_{0} \cos \theta $$

$$ v_{y 0}=v_{0} \sin \theta $$

任意时刻t:

$$ x(t)=v_{x0}t $$

$$ y(t)=v_{y0}t-\frac{1}{2} gt^{2} $$

$$ v_{y}(t)=v_{y0}-gt $$

````python
import math

class ProjectileMotion:
    def __init__(self,v0,theta):
        self.v0=v0
        self.theta=math.radians(theta)
        self.vx=self.v0 * math.cos(self.theta)
        self.vy0=self.v0 * math.sin(self.theta)
        self.g = 9.8
    # 计算x和y坐标以及y方向的速度
    def x(self, t):
        return self.vx * t
    def y(self, t):
        return self.vy0 * t - 0.5 * self.g * (t**2)
    def vy(self, t):
        return self.vy0 - self.g * t
    # 生成表格
    def generate_table(self, t_end, dt):
        table=[]
        t = 0
        table.append((0,0,0,self.vy0))
        while t <= t_end and self.y(t) >= 0:
            table.append((t,self.x(t),self.y(t),self.vy(t)))
            t += dt
        return table
    # 显示表格
    def show_table(self, t_end, dt):
        table = self.generate_table(t_end,dt)
        for t,x,y,vy in table:
            print(f"t = {t}, x = {x}, y = {y}, vy = {vy}")
    # 寻找最大高度
    def max_height(self, t_end, dt):
        table = self.generate_table(t_end, dt)
        current_max = table[0][2]
        for item in table:
            if item[2] > current_max:
                current_max = item[2]
        return current_max


pm = ProjectileMotion(float(input()),float(input()))
pm.show_table(float(input()),float(input()))
print(pm.max_height(float(input()),float(input())))
````

题目：牛顿冷却模型

定义一个类CoolingModel，模拟物体降温过程。

已知：
- 初始温度 $T_0$
- 环境温度 $T_{\text{env}}$
- 冷却系数 $k$
- 时间步长 $dt$

牛顿冷却模型:
$$\frac{dT}{dt}=-k(T-T_{env})$$

离散更新公式：
$$ T_{\text{new}} = T - k(T - T_{\text{env}})dt $$

````python
import math
class CoolingModel:
    def __init__(self,T0,T_env,k):
        self.T0 = T0
        self.T_env = T_env
        self.k = k
    # 模拟冷却过程
    def simulate(self, dt, threshold):
        steps = 0 # 记录迭代次数
        data = [(0.0, self.T0)]
        T = self.T0
        # 迭代直到温度接近环境温度
        while abs(T-self.T_env) > threshold:
            T = T - self.k * (T - self.T_env) * dt
            data.append((data[-1][0] + dt, T))# data[-1][0]是上一个时间点，dt是时间增量
            steps += 1
        print(f"steps = {steps}")
        return data
    # 显示数据
    def show_data(self, dt, threshold):
        # 获取模拟数据
        data = self.simulate(dt, threshold)
        # 打印时间和温度
        for t, T in data:
            print(f"t = {t}, T = {T}")
cm = CoolingModel(float(input()), float(input()), float(input()))
cm.show_data(float(input()), float(input())) 
````

# 第45题

题目：银行账户复利增长模型

定义一个类 BankAccount，模拟账户资金按固定利率增长。

已知

- 初始金额 P0

- 年利率 r

- 每次计息的时间步长 dt （单位：年）

离散更新公式：

$$ P_{\text {new }}=P(1+r \cdot d t) $$

````python
class Bankaccount:
    # 初始化账户，设置初始金额和年利率
    def __init__(self, P0, r):# P0是初始金额，r是年利率
        self.P0 = P0
        self.r = r
    def simulate(self, dt, target):
        steps = 0 # 记录迭代次数
        t = 0
        data = [(0.0, self.P0)]
        P = self.P0
        while P < target:
            P = P + self.r * P * dt
            t += dt
            data.append((t, P))
            steps += 1
        return data, steps
    def show_data(self, dt, target):
        data = self.simulate(dt, target)[0]
        steps = self.simulate(dt, target)[1]
        for t, P in data:
            print(f"t = {t}, P = {P}")
        print(f"steps = {steps}")
account = Bankaccount(float(input()), float(input()))
account.show_data(float(input()), float(input()))
````
