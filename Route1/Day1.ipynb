# 回答你的问题：`range(某个整数)` 到底是不是创建了一个连续整数列表？

## 简短答案
**不完全是。**

更准确地说：

> `range(某个整数)` 会生成一个“表示连续整数序列的范围对象”，它通常表示从 **0** 开始，到这个整数前一个为止。

比如：

```python

```

表示的是：

```python
0, 1, 2, 3, 4
```

**不是**：

```python
1, 2, 3, 4, 5
```

# 重点 1：`range(5)` 不是从 1 开始，而是从 0 开始

例如：

```python
for i in range(5):
    print(i)
```

输出：

```text
0
1
2
3
4
```

所以：

```python
range(n)
```

表示的是：

> 从 `0` 到 `n-1`

---

# 重点 2：它通常不把自己当成真正的列表来用
在初学阶段，你可以先把它理解成：

> “一个可以按顺序给出若干整数的东西”

例如：

```python
range(5)
```

你可以在循环里直接用它：

```python
for i in range(5):
    print(i)
```

如果你真想把它变成列表，可以写：

```python
print(list(range(5)))
```

输出：

```python
[0, 1, 2, 3, 4]
```

所以更准确的说法是：

- `range(5)` **不是字面上直接写出来的列表**
- 但它可以表示一个这样的整数序列
- 必要时可以转成列表

# 常见查找思路
比如要找某个数 `target` 是否在列表里，或者找它第一次出现的位置：

found = -1

这里的逻辑是：

- 先假设没找到，令 `found = -1`
- 从前往后检查每个元素
- 一旦找到，就把下标记下来
- 然后 `break` 结束循环

这是一种常见技巧。
