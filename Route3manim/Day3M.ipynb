题目：创世纪——行星的诞生与蜕变

为了呼应你的引力轨道项目，我们要用 Manim 绘制一个“行星观测窗口”。

任务要求
定义一个类 PlanetIntro(Scene)：

## 绘制行星：

创建一个半径为 1.5 的圆（Circle），填充颜色为蓝色（BLUE），填充透明度（fill_opacity）设为 0.5。

在圆的中心添加文字 "Planet"（Text）。

## 动画编排：

第一步：使用 Write 动画显示文字，同时使用 Create 动画画出圆。

第二步：等待 1 秒（self.wait(1)）。

第三步：形态蜕变。将这个圆和文字整体，通过 ReplacementTransform 变成一个黄色的正方形（Square），边长为 3，且内部文字变为 "Star"。

## 布局细节：

所有物体初始位置应在屏幕中心。

```python
from manim import *
```

```python
%%manim -qm -v WARNING PlanetIntro
from manim import *

class PlanetIntro(Scene):
    def construct(self):
        circle = Circle(radius=1.5, color=BLUE, fill_opacity=0.5)
        planet_text = Text("Planet").move_to(circle.get_center())
        planet_group = VGroup(circle, planet_text)

        square = Square(side_length=3, color=YELLOW)
        star_text = Text("Star").move_to(square.get_center())
        star_group = VGroup(square, star_text)

        self.play(
            Create(circle),
            Write(planet_text)
        )
        self.wait(1)
        self.play(
            ReplacementTransform(planet_group, star_group)
        )
        self.wait(1)
```
